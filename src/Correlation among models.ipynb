```python
import torch
import matplotlib.pyplot as plt
from MLP_trainer import MLP
import pandas as pd
import numpy as np
```

```python
input_size = 784
batch_size = 64
hidden_size = 100
num_classes = 10
learning_rate = 0.001
num_epochs = 10
model_num=0
```

```python
model = MLP(input_size=input_size, hidden_size=hidden_size, num_classes=num_classes)
layers = []
for i in range(30):
    model.load_state_dict(torch.load(f'../model/model_{i}.ckpt'))
    layers.append(model.fc1.weight.data.numpy().flatten())

array = np.vstack(layers)
cov_matrix = np.corrcoef(array)
```

```python
max_corr = 0
for i in range(30):
    for j in range(i+1, 30):
        max_corr = max(max_corr, cov_matrix[i, j])
max_corr
```
